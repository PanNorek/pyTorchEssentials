```python
import numpy as np
import torch
import torch.nn as nn
```

```python
widenet = nn.Sequential(
    nn.Linear(2,4),  
    nn.Linear(4,3),  
    )


deepnet = nn.Sequential(
    nn.Linear(2,2),  
    nn.Linear(2,2),  
    nn.Linear(2,3),  
    )


print(widenet)
print('\n')
print(deepnet)
```
